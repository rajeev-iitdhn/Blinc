```
import pandas as pd
from random import randint
import random
import numpy as np
from sklearn.utils import shuffle
```

```
columns  = ['anony_id','user_id','type_of_user','origin','destination','no_of_adults','no_children','class','trip_type',
			'no_of_infants','dep_of_date','return_date','search_timestamp']
```

```
type_of_user = ['FL','FNL','RL','RNL']
```

```
airport_code = ['BOM','BLR','DEL','HYD','MAA','CCU','AMD','PNQ',
				'GAU','LKO','JAI','BBI','PAT','COK','TRV','CCJ','CNN']
```

```
trip_type = ['OW','RT']
passenger_class = ['ECO','PECO','BSN','FCLASS']
```

```
sample_data  = pd.DataFrame(columns=columns)

for i in range(500):
    user_id = i
    #sample_data.loc[i,"user_id"]=str("user"+str(i))
    random.shuffle(airport_code)
    random.shuffle(type_of_user)
    random.shuffle(passenger_class)
    sample_data.loc[i,'type_of_user']=type_of_user[randint(0,len(type_of_user)-1)]
    sample_data.loc[i,'origin'] =airport_code[0]
    sample_data.loc[i,'destination'] = airport_code[1]
    #sample_data.loc[i,'type_of_module']='Search'
    sample_data.loc[i,'no_of_adults']=int(randint(1,2))
    sample_data.loc[i,'no_of_infants']=int(randint(0,2))
    sample_data.loc[i,'no_children']=int(randint(0,2))
    sample_data.loc[i,'class']=passenger_class[randint(0,len(passenger_class)-1)]
    sample_data.loc[i,'trip_type']=trip_type[randint(0,len(trip_type)-1)]
    sample_data.loc[i,'no_of_infants'] = int(randint(0,2)) 
    if (sample_data.loc[i,'type_of_user']=='RNL' or sample_data.loc[i,'type_of_user']=='FNL'):
      sample_data.loc[i,'user_id']=''
      sample_data.loc[i,'anony_id']=str("anony"+str(i))
    else:
      sample_data.loc[i,"user_id"]=str("user"+str(i))
      sample_data.loc[i,'anony_id']=""
    if (sample_data.loc[i,'trip_type']=="OW"):
        dep_day = randint(1,20)
        dep_month = randint(1,12)
        dep_year = 2020

        sample_data.loc[i,'dep_of_date']= "{}-{}-{}".format(dep_day,dep_month,dep_year)
        sample_data.loc[i,'return_date']= ''
        sample_data.loc[i,'search_timestamp'] = pd.Timestamp(year = 2020,  month = 1, day = randint(1,7),
                                                             hour=randint(0,23),
                                                             second=randint(1,59)) 
        #pd.to_datetime(sample_data.loc[i,'dep_of_date'])
    else:
        dep_day = randint(1,20)
        dep_month = randint(1,12)
        dep_year = 2020
        #dep_format_date =
        sample_data.loc[i,'dep_of_date']= "{}-{}-{}".format(dep_day,dep_month,dep_year)
        return_date = randint(dep_day,20)
        return_month = randint(dep_month,12)
        return_year = 2020
        sample_data.loc[i,'return_date']= "{}-{}-{}".format(return_date,return_month,return_year)
        sample_data.loc[i,'search_timestamp'] = pd.Timestamp(year = 2020,  month = 1, day = randint(1,7),
                                                             hour=randint(0,23),
                                                             second=randint(1,59))

sample_data['dep_of_date']=pd.to_datetime(sample_data['dep_of_date'])
sample_data['return_date']=pd.to_datetime(sample_data['return_date'])
sample_data['no_children']=sample_data['no_children'].astype(int)
sample_data['no_of_infants']=sample_data['no_of_infants'].astype(int)
sample_data['no_of_adults']=sample_data['no_of_adults'].astype(int)
sample_data['search_timestamp']=pd.to_datetime(sample_data['search_timestamp'])


```
