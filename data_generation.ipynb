```
type_of_user = ['FL','FNL','RL','RNL']
airport_code = ['BOM','BLR','DEL','HYD','MAA','CCU','AMD','PNQ','GAU','LKO','JAI','BBI','PAT','COK','TRV','CCJ','CNN']
trip_type = ['OW','RT']
passenger_class = ['ECO','PECO','BSN','FCLASS']
travel_type=["dom","int"]
departure_month=["jan","feb","mar","apr","may","jun","july","aug","sep","oct","nov","dec"]
departure_day=["sunday","monday","tuesday","wednesday","thrusday","friday","saturday"]
departure_time=["mor","afternoon","evening","night"]
travel_during_public_holiday=["yes","no"]
travel_during_weekend=["yes","no"]
travel_during_peak_seasons=["yes","no"]
airlines=["indigo","sjet","airindia","jetairways","airasia","gulfair"]
lowest_price_selected=["yes","no"]
lowest_timetaken_selected=["yes","no"]


cols=["id","user_id","anonymous_user_id","user_type","origin","destination","trip_type","class","number_of_adults","number_of_children","number_of_infants","departure_date",
         "return_date","travel_type","departure_month","departure_day","departure_time","return_month","return_day","return_time","travel_during_public_holiday","travel_during_weekend"]
```

```
now=pd.to_datetime(datetime.datetime.now().strftime("%Y/%m/%d %H:%M:%S"))
dummy_data=pd.DataFrame(columns=cols)
for i in range(1,1000):
  random.shuffle(airport_code)
  random.shuffle(type_of_user)
  random.shuffle(passenger_class)
  dummy_data.loc[i,'number_of_adults']=np.int(randint(1,2))
  dummy_data.loc[i,'number_of_infants']=np.int(randint(0,2))
  dummy_data.loc[i,'number_of_children']=np.int(randint(0,2))
  dummy_data.loc[i, "user_type"]=shuffle(type_of_user)[0]
  dummy_data.loc[i, "id"]=int(i)
  if (dummy_data.loc[i,'user_type']=='RNL' or dummy_data.loc[i,'user_type']=='FNL'):
      dummy_data.loc[i,'user_id']=''
      dummy_data.loc[i,'anonymous_user_id']=str("anony"+str(randint(1,100)))
  
  else:
      dummy_data.loc[i,"user_id"]=str("user"+str(randint(1,100)))
      dummy_data.loc[i,'anonymous_user_id']=""

  if (dummy_data.loc[i,'trip_type']=="OW"):
      dep_day = randint(1,20)
      dep_month = randint(1,12)
      dep_year = 2020
      dummy_data.loc[i,'departure_date']= "{}-{}-{}".format(dep_day,dep_month,dep_year)
      dummy_data.loc[i,'return_date']= ''
      #pd.to_datetime(sample_data.loc[i,'dep_of_date'])
  else:
      dep_day = randint(1,20)
      dep_month = randint(1,12)
      dep_year = 2020
      #dep_format_date =
      dummy_data.loc[i,'departure_date']= "{}-{}-{}".format(dep_day,dep_month,dep_year)
      return_date = randint(dep_day,20)
      return_month = randint(dep_month,12)
      return_year = 2020
      dummy_data.loc[i,'return_date']= "{}-{}-{}".format(return_date,return_month,return_year)
  
  dummy_data['departure_date']=pd.to_datetime(dummy_data['departure_date'])
  dummy_data['return_date']=pd.to_datetime(dummy_data['return_date'])
  dummy_data.loc[i,'origin'] =airport_code[0]
  dummy_data.loc[i,'destination'] = airport_code[-1]
  dummy_data.loc[i,'trip_type']=trip_type[randint(0,len(trip_type)-1)]
  dummy_data.loc[i,'class']=passenger_class[randint(0,len(passenger_class)-1)]
  dummy_data.loc[i,"travel_type"]=shuffle(travel_type)[0]
  dummy_data.loc[i, "departure_month"]=shuffle(departure_month)[0]
  dummy_data.loc[i, "departure_day"]=shuffle(departure_day)[0]
  dummy_data.loc[i, "departure_time"]=shuffle(departure_time)[0]
  dummy_data.loc[i, "return_month"]=shuffle(departure_month)[1]
  dummy_data.loc[i, "return_day"]=shuffle(departure_day)[1]
  dummy_data.loc[i, "return_time"]=shuffle(departure_time)[1]
  dummy_data.loc[i, "travel_during_public_holiday"]=shuffle(travel_during_public_holiday)[1]
  dummy_data.loc[i, "travel_during_weekend"]=shuffle(travel_during_weekend)[1]
  dummy_data.loc[i, "travel_during_peak_seasons"]=shuffle(travel_during_peak_seasons)[1]
  dummy_data.loc[i,"search_time"]=pd.Timestamp(2020, 1, randint(15,22), randint(1,23))
  dummy_data.loc[i,"airline_selected"]=shuffle(airlines)[0]
  dummy_data.loc[i,"price_selected"]=randint(3000,9000)
  dummy_data.loc[i, "lowest_price_selected"]=shuffle(lowest_price_selected)[0]
  dummy_data.loc[i,"lowest_timetaken_selected"]=shuffle(lowest_timetaken_selected)[0]



dummy_data["hours_from_now"]=dummy_data.search_time.apply(lambda x: (now-x).days*24+int((now-x).seconds/3600))
dummy_data[["id","number_of_adults","number_of_children","number_of_infants"]]=dummy_data[["id","number_of_adults","number_of_children","number_of_infants"]].astype("int")
```

```
input_call=dummy_data.iloc[12:18,3:-1]
# m=len(input_call)
mp=list(input_call.index)
m=len(list(input_call.index))
def recommend(m):
  output=[]
  indices=[]
  for i in mp:
    indices.append(i)
    tester=list(input_call.loc[i,:].values)
    ush=dummy_data.iloc[1:,3:-1]

    ind=[]
    matched=[]
    for k in list(ush.index):
      matched.append(len(set(list(ush.loc[k,:].values)) & set(tester)))
      ind.append(k)

    df=pd.DataFrame()
    df["match"]=pd.Series(matched)
    df["ind"]=pd.Series(ind)
    toppers=df.nlargest(int(np.ceil(0.1*len(ush))),"match")

    dk=pd.crosstab(ush.loc[toppers.ind,:].lowest_price_selected, ush.loc[toppers.ind,:].lowest_timetaken_selected)

    if np.where(dk==np.max(np.matrix(dk)))[0][0]==0 and np.where(dk==np.max(np.matrix(dk)))[1][0]==0:
      output.append("NO ONE")
    elif np.where(dk==np.max(np.matrix(dk)))[0][0]==1 and np.where(dk==np.max(np.matrix(dk)))[1][0]==1:
      output.append("LP & LT")
    elif np.where(dk==np.max(np.matrix(dk)))[0][0]==0 and np.where(dk==np.max(np.matrix(dk)))[1][0]==1:
      output.append("LT")
    elif np.where(dk==np.max(np.matrix(dk)))[0][0]==1 and np.where(dk==np.max(np.matrix(dk)))[1][0]==0:
      output.append("LP")
  return (pd.DataFrame(zip(indices,output),columns=["Index","Recommended_filter"]))
recommend(m)
```

```
dum=pd.DataFrame()
dum["day"]=pd.Series([i+1 for i in range(15)])
dum["target"]=pd.Series([10,10,10,10,10,10,10,10,10,10,10,10,10,10,10])
dum["achievement"]=pd.Series([8,9,6,14,9,12,14,9,7,9,8,7,5,6,12])
dum["%_achievement"]=pd.Series(dum.achievement*100/dum.target).apply(int)
# dum["lag"]=dum.target-dum.achievement
dum["shifted_value"]=pd.Series(dum["%_achievement"].shift(1))
dum=dum.fillna(0)
dum["%_rf"]=pd.Series((-dum.shifted_value+dum["%_achievement"])*100/(dum.shifted_value))[1:]
dum=dum.fillna(0)
dum["%_rf"]=dum["%_rf"].apply(int)
dum=dum.fillna(0)
dum["cons_till_today"]=pd.Series(dum["%_rf"].cumsum()/dum.day).apply(int)
dum["total_score"]=pd.Series(dum["%_achievement"]*0.1+dum.cons_till_today*0.7).apply(int)
```
